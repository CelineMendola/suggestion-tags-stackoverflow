### Most frequent words is posts

```python
from wordcloud import WordCloud
```

```python
from collections import Counter
```

```python
def to_text(a):
    ''' Concatenate all the text data of the a serie'''
    text = ''
    for i in range(len(a)):
        text += ' ' + a[i].strip()
    return text
```

```python
all_body = to_text(df['Body'])
```

```python
# Most common words in body
c = Counter(all_body.split())
c.most_common(10)
```

### Tags

```python
df['Tags'] = df['Tags'].apply(clean_tags)
```

```python
#tags
tags=[]
for i in range(df.shape[0]):
    tags+=df.iloc[i,df.columns.get_loc('Tags')]
```

```python
tags_counts = Counter(tags)
tags_counts.most_common(50)
```

```python
len(tags)
```

```python
dh = pd.DataFrame(tags_counts.most_common(30), columns = ['Tag', 'Count'])
dh.plot.bar(x='Tag', y='Count')
plt.title('Most common Tags', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
tags_text = to_text(tags)
```

```python
# Word cloud of tags

tags_text = to_text(tags)
plt.figure(figsize=(40, 30))
wordcloud = WordCloud(width=3000,
                      height=2000,
                      background_color = 'white',
                      collocations=False,
                      random_state=0,
                      stopwords=[]).generate(tags_text)
plt.imshow(wordcloud)
plt.title('Most frequent tags', fontsize=50)
plt.axis("off")
plt.show()
```

## II. Preprocessing and data analysis

### Remove stopwords and  lemmatize text

```python
import gensim, spacy
import gensim.corpora as corpora
from nltk.corpus import stopwords
```

```python
df['Body_clean'] = df['Body'].apply(remove_stop_words)
df['Title_clean'] = df['Title'].apply(remove_stop_words)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
df['Body_clean'] = df['Body_clean'].apply(lemmatize)
df['Title_clean'] = df['Title_clean'].apply(lemmatize)
```

```python
#Most common words in bodies

from pylab import rcParams
rcParams['figure.figsize'] = 20, 10

c = Counter(to_text(df['Body_clean']).split())
lst = c.most_common(30)
dh = pd.DataFrame(lst, columns = ['Word', 'Count'])
dh.plot.bar(x='Word', y='Count')
plt.title('Most common words in posts bodies', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```

```python
#pattern = r'\b[abdefghijklmnopqstuwxxyz]\b'
#df['Body_clean'] = df['Body_clean'].apply(lambda x: re.sub(pattern,' ',x))
#df['Title_clean'] = df['Title_clean'].apply(lambda x: re.sub(pattern,' ',x))
```

```python
# Most common words in titles
c = Counter(to_text(df['Title_clean']).split())
lst = c.most_common(30)
dh = pd.DataFrame(lst, columns = ['Word', 'Count'])
dh.plot.bar(x='Word', y='Count')
plt.title('Most common words in posts titles', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
