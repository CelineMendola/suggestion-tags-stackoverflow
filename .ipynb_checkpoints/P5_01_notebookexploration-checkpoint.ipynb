## II. Preprocessing and data analysis

### Remove stopwords and  lemmatize text

```python
import gensim, spacy
import gensim.corpora as corpora
from nltk.corpus import stopwords
```

```python
# Load NLTK stopwords
stop_words = stopwords.words('english')

# add stopwords
import string

letters = [l for l in string.ascii_lowercase]
letters.remove('r')
letters.remove('c')
stop_words.extend(['+', '-','.+','├']) #+letters)
```

```python
def remove_stop_words(text):
    '''remove stopwords in text data'''
    lis=[]
    for token in text.split():
        if (token not in stop_words) and (token not in gensim.parsing.preprocessing.STOPWORDS):
            lis.append(token)
    return(' '.join(lis))
```

```python
df['Body_clean'] = df['Body'].apply(remove_stop_words)
df['Title_clean'] = df['Title'].apply(remove_stop_words)
```

```python
import spacy
nlp = spacy.load("en_core_web_sm")
```

```python
def lemmatize(text):
    '''lemmatize text data'''
    doc = nlp(text)
    tokens = [token.lemma_ for token in doc if not (token.is_stop or token.is_punct)]
    return ' '.join(tokens)
```

```python
df['Body_clean'] = df['Body_clean'].apply(lemmatize)
df['Title_clean'] = df['Title_clean'].apply(lemmatize)
```

```python
#Most common words in bodies

from pylab import rcParams
rcParams['figure.figsize'] = 20, 10

c = Counter(to_text(df['Body_clean']).split())
lst = c.most_common(30)
dh = pd.DataFrame(lst, columns = ['Word', 'Count'])
dh.plot.bar(x='Word', y='Count')
plt.title('Most common words in posts bodies', fontsize=20)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.show()
```
